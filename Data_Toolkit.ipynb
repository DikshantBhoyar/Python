**2.  How does broadcasting work in NumPy?**
  - Broadcasting in NumPy allows operations on arrays of different shapes without explicitly reshaping them. Instead of manually expanding smaller arrays, NumPy automatically adjusts their shapes to make operations efficient.  

   - Rules of Broadcasting

       - If an array has fewer dimensions than another, it is prepended with ones in its shape to match the larger array.  
       - The size in each dimension of the output is the maximum of the input sizes in that dimension.  
       - An array can be used in calculations if its size in a particular dimension matches the output size or is exactly 1.  
       - If an array has a dimension size of 1, the first data entry in that dimension is used for all calculations along that dimension.  

   - When Are Arrays Broadcastable?

      - A set of arrays can be broadcasted if:  
          - They have the same shape.  
          - They have the same number of dimensions, and the length of each dimension is either the same or 1.  
          - If an array has fewer dimensions, it can be prepended with ones to satisfy the above conditions.  

            
            Example 1: Adding a Scalar to an Array

                             import numpy as np
                             arr = np.array([[1, 2, 3], [4, 5, 6]])  # Shape (2,3)
                             result = arr + 10  # Scalar (Shape ()) is broadcasted

                             print(result)

             Output:

                    [[11 12 13]  
                      [14 15 16]]


                -----------------------------------------------------


          Example 2: Broadcasting a 1D Array to a 2D Array

                      A = np.array([[1, 2, 3], [4, 5, 6]])  # Shape (2,3)
                      B = np.array([10, 20, 30])  # Shape (1,3), gets broadcasted

                      result = A + B  
                      print(result)

          Output:

                 [[11 22 33]  
                   [14 25 36]]

  - Explanation:
    - B has shape (1,3), while A has (2,3).  
    - B is broadcasted to (2,3) by repeating its values along the first axis.

**4.  Explain the use of the groupby() method in Pandas?**
   - The "groupby()" method in Pandas is used to "group data based on one or more columns" and perform "aggregate operations" like sum, mean, count, etc. It is similar to the "GROUP BY" clause in SQL.  

    - Why Use " groupby()"?

         I. Helps in analyzing large datasets by grouping similar data.  
II. Allows applying aggregate functions like sum, mean, max, min, and count.  
III. Useful for summarizing data efficiently.  



          Example 1: Grouping by a Single Column

                        import pandas as pd  

                        data = {
                           "Department": ["HR", "IT", "HR", "IT", "Finance"],
                           "Salary": [50000, 70000, 52000, 75000, 60000]
                         }

                        df = pd.DataFrame(data)

                        Grouping by 'Department' and calculating the average salary
                        result = df.groupby("Department")["Salary"].mean()

                        print(result)

                        
              Output:

                        Department
                          Finance    60000.0
                          HR         51000.0
                          IT         72500.0
                        Name: Salary, dtype: float64
                       
                          ---------------------

              Example 2: Grouping by Multiple Columns

                       data = {
                            "Department": ["HR", "IT", "HR", "IT", "Finance", "IT"],
                            "Gender": ["M", "F", "F", "M", "M", "F"],
                            "Salary": [50000, 70000, 52000, 75000, 60000, 72000]
                       }

                       df = pd.DataFrame(data)


                       ---------------------

- Grouping by 'Department' and 'Gender' and calculating the sum of salaries

                        result = df.groupby(["Department", "Gender"])["Salary"].sum()

                        print(result)

                  Output:

                         Department  Gender
                         Finance     M         60000
                         HR          F        102000
                         IT          F        142000
                                     M         75000
                         Name: Salary, dtype: int64

                         --------------------------



- Common Aggregation Functions with "groupby()"
    - ".sum()" → Total sum  
    - ".mean()" → Average value  
    - ".count()" → Number of occurrences  
    - ".max()" / ".min()" → Maximum / Minimum value  
    - ".agg(['sum', 'mean'])" → Apply multiple functions

| **Library Requirement** | Requires importing NumPy (`import numpy as np`) | Built into Python, no need for an external library |

**8. What does the term “vectorized operation” mean in NumPy?**  
- A vectorized operation in NumPy refers to performing operations on entire arrays without using explicit loops. NumPy applies the operation element-wise, making computations faster and more efficient by leveraging low-level optimizations.  

   - Uses of Vectorized Operations :  

     - I. Faster Execution :
          - NumPy operations run much faster than equivalent Python loops because they use compiled code.  

     - II. Memory Efficiency :  
          - Reduces memory overhead by performing operations directly on arrays instead of creating intermediate variables.  

     - III. Concise and Readable Code :  
          - Eliminates the need for `for` loops, making the code shorter and easier to understand.  

     - IV. Parallel Processing :
          - Utilizes optimized CPU instructions for efficient execution.  



- Common Vectorized Operations in NumPy  

| Operation  | Example |
|------------|---------|
| Addition   | `arr1 + arr2` |
| Subtraction | `arr1 - arr2` |
| Multiplication | `arr * 2` |
| Division | `arr / 2` |
| Square Root | `np.sqrt(arr)` |
| Exponentiation | `np.exp(arr)` |

**13. Why is handling missing data important in Pandas?**
   -  Handling missing data is crucial because it directly impacts the "accuracy and reliability" of data analysis and model building. Incomplete or missing data can lead to incorrect interpretations, biased results, and unreliable models. Pandas provides multiple ways to handle missing data, making it a key part of data preprocessing.

- Handling Missing Data is Important :  

 - Ensures Accurate Analysis :
      - Missing values can distort the analysis. For example, if the mean or median is calculated with missing values, it may give a skewed result.  

 - Improves Model Performance :
      - Machine learning models typically require complete datasets. Missing data can cause errors or lead to inaccurate predictions, affecting model performance. Handling missing data properly can enhance "model accuracy" and "prediction reliability".  

 - Prevents Bias :
      - If the missing values are not handled, they may introduce bias into the results. For example, if only certain rows with missing data are removed, it could skew the representation of the data.  

 - Consistency in Data :
      - Handling missing data ensures the "consistency" and "completeness" of the dataset, making it easier to work with and reducing the chances of data integrity issues later in the analysis.  
 - Improves Visualization :
      - Visualizations might be distorted if missing values are not handled properly. For instance, if some values are missing in a time series, the plot might look incomplete or have gaps.  

- Methods for Handling Missing Data in Pandas :

   - Removing Missing Data  
       - Use `dropna()` to remove rows or columns with missing values.  

                    df.dropna()  # Removes rows with any missing value

   - Filling Missing Data
       - Use `fillna()` to replace missing values with a specific value, the mean, median, or any other method.  

                    df.fillna(value=0)  # Replace missing values with 0
                    df.fillna(df.mean())  # Replace missing values with column mean

  - Forward Fill and Backward Fill  
       - Use `ffill()` to propagate the next valid observation or `bfill()` to propagate the previous valid observation.  

                    df.ffill()  # Forward fill missing values
                    df.bfill()  # Backward fill missing values

 - Interpolating Missing Values
       - Use `interpolate()` to estimate missing values based on other available values (commonly used in time series).  

                    df.interpolate()  # Linear interpolation to fill missing values

**15.  How does NumPy handle multidimensional arrays?**
  - NumPy provides a powerful, efficient way to work with "multidimensional arrays". These arrays are fundamental for representing data in fields such as data science, machine learning, and scientific computing.

  - Aspects of NumPy's Multidimensional Arrays :

     - Creation and Structure :
- NumPy arrays can have any number of dimensions. A 1D array is a simple list, while a 2D array is a matrix with rows and columns. For 3D arrays and beyond, NumPy uses higher dimensions, often referred to as a "tensor" structure.

     - Shape and Size :  
- The "shape" of a NumPy array refers to the number of elements in each dimension (e.g., a matrix with 3 rows and 4 columns has a shape of (3, 4)).
- "ndim" specifies the number of dimensions an array has, making it easy to determine whether the array is 1D, 2D, or higher.

     - Indexing and Slicing :
- Elements in multidimensional arrays can be accessed through "indexing" for each dimension. NumPy supports advanced "slicing" to select specific portions of the array, which is useful for manipulating subsets of data.

     - Operations on Multidimensional Arrays :  
- NumPy enables "vectorized operations" on entire arrays. Mathematical operations like addition, subtraction, multiplication, etc., can be applied to arrays, including those with multiple dimensions, without needing explicit loops.

     - Reshaping Arrays :
- The "reshape()" function in NumPy allows you to modify the structure of an array, such as transforming a 1D array into a 2D or 3D array, as long as the total number of elements remains the same.

     - Broadcasting :  
- Broadcasting is a powerful feature that allows NumPy to perform operations on arrays of different shapes. Smaller arrays are "broadcast" to match the shape of the larger array, enabling element-wise operations without explicit replication of data.

**17. Explain the difference between apply() and map() in Pandas?**

--->


| **Feature**             | **`apply()`**                              | **`map()`**                                  |
|-------------------------|--------------------------------------------|----------------------------------------------|
| **Applies To**           | Series and DataFrame                       | Series only                                  |
| **Functionality**        | Can apply functions across rows, columns, or the entire DataFrame | Element-wise operations                      |
| **Return Type**          | Varies: Series, DataFrame, or scalar       | Always returns a Series                      |
| **Complexity**           | Can handle complex operations and return different data structures | Simpler, used mainly for element-wise mapping |
| **Performance**          | Slower for simple element-wise operations  | Faster for element-wise operations           |
| **Use Case**             | Complex transformations across rows or columns | Simple, element-wise transformations (e.g., mapping values, replacing) |
| **Flexibility**          | More flexible, supports multiple types of operations | Less flexible, limited to element-wise operations |
| **Axis Support**         | Supports axis parameter (rows or columns)  | No axis support, works only on individual elements of the Series |
| **Aggregation**          | Can be used for aggregation functions like `sum()`, `mean()`, etc. | Not used for aggregation                     |

---
          - Example for "apply()" (on a DataFrame) :


                     import pandas as pd

                     df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})  # Create a DataFrame

                     result = df.apply(lambda x: x.sum(), axis=0)   # Use apply() to sum the values in each column (axis=0)
                     print(result)


         Output:

                    A     6
                    B    15
                    dtype: int64

         In this example, "apply()" is used to sum the values in each column ("axis=0").

                      --------------------------------

         - Example for `map()`** (on a Series) :

                   import pandas as pd

                   s = pd.Series(['aditi', 'sannya', 'surya'])  # Create a Series

                   result = s.map({'aditi': 'dikshant', 'sannya': 'archana'})  # Use map() to replace values using a dictionary
                   print(result)
         Output:

                    0    dikshant
                    1    archana
                    2      NaN
                 dtype: object


        In this example, "map()" is used to replace values in the Series based on a dictionary. "aditi" is replaced with "dikshant" and "sannya" is replaced with "archana". "surya" is not in the dictionary, so it becomes "NaN".

**21. Why is NumPy’s array slicing faster than Python’s list slicing?**
- NumPy’s array slicing is faster than Python’s list slicing due to:  

     - Memory Efficiency – NumPy arrays use contiguous memory blocks, making access and slicing operations faster.  
     - No Data Copying – Slicing in NumPy creates a view, reducing memory usage and processing time.  
     - Vectorized Operations – NumPy uses low-level optimizations, speeding up computations.  
     - Efficient Indexing – NumPy arrays support advanced indexing techniques for better performance.  

                - Example: NumPy vs. List Slicing

                         import numpy as np
                         import time

                         arr = np.arange(1000000)

                         start = time.time()
                         slice_arr = arr[:500000]
                         end = time.time()
                         print("NumPy Slicing Time:", end - start)

                         lst = list(range(1000000))

                         start = time.time()
                         slice_lst = lst[:500000]
                         end = time.time()
                         print("List Slicing Time:", end - start)

              Output :

                       "NumPy Slicing Time": 0.0001s  
                       "List Slicing Time": 0.002s  

              NumPy slicing is faster due to optimized memory management and low-level operations.
