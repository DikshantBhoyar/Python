**1. What is the difference between a function and a method in Python?**
   - The main difference between a function and a method** in Python lies in their association with objects:

**Function**  :
- A function is an independent block of code that performs a specific task.  
- It is defined using the `def` keyword.  
- It can be called anywhere in the program.  
- Functions can be built-in (like `len()`, `print()`) or user-defined.  

      Example :

      def greet(name):
          return f"Hello, {name}!"
     
      print(greet("dikshant"))  # Output: Hello, dikshant!

  **Method**  :
- A method is similar to a function but is associated with an object.  
- It is defined inside a class and operates on instances of that class.  
- It is called using the dot (`.`) notation on an object.  
- Methods can modify object attributes or perform operations related to the object.  


     Example :

        class Person:
             def __init__(self, name):
                   self.name = name

        def greet(self):
             return f"Hello, {self.name}!"

        p = Person("dikshant")
        print(p.greet())  #Output : Hello, dikshant!







| Feature  | Function  | Method  |
|----------|----------|---------|
| **Definition** | Defined using `def` outside a class | Defined inside a class |
| **Association** | Not associated with any object | Associated with an object |
| **Calling Style** | Called directly (`greet("dikshant")`) | Called on an object (`p.greet()`) |
| **Self Parameter** | Does not require `self` | Requires `self` to access object attributes |

**2. Explain the concept of function arguments and parameters in Python?**
   - In Python, when defining a function, we use **parameters**, and when calling a function, we pass **arguments**.  

        - **Parameters** are variables that are used in the function definition to accept input values.  
        - **Arguments** are actual values passed to a function when calling it.

.

 Example:

        def greet(name):  # "name" is a parameter
            print(f"Hello, {name}!")

        greet("dikshant")  # "dikshant" is an argument


.

Types of Function Arguments in Python :

Python supports multiple ways to pass arguments:

   - Positional Arguments :
- These are passed in the same order as the function parameters.  
- The number of arguments must match the number of parameters.

Example:

     def add(a, b):
         return a + b

    print(add(5, 3))  # Output: 8
    # Here, `5` is assigned to `a`, and `3` is assigned to `b` based on position.



  - Keyword Arguments :
- Arguments are passed with parameter names, so the order doesn’t matter.

Example:

     def introduce(name, age):
          print(f"My name is {name} and I am {age} years old.")

     introduce(age=23, name="Dikshant") #Here, even though we changed the order, Python correctly assigns values.



  - Default Arguments :
- Parameters can have default values, which are used if no argument is provided.

Example:

      def greet(name="aditi"):
          print(f"Hello, {name}!")

      greet()  # Output: Hello, aditi!
      greet("dikshant")  # Output: Hello, dikshant!
      # If no argument is passed, `"aditi"` is used as the default.



- Variable-Length Arguments :

    - Python allows functions to accept a variable number of arguments using *args(for positional arguments) and **kwargs (for keyword arguments).

Using *args (Arbitrary Positional Arguments)
- Allows passing multiple arguments as a tuple.


           def sum_all(*numbers):
               return sum(numbers)

           print(sum_all(1, 2, 3, 4))  # Output: 10



Using *kwargs (Arbitrary Keyword Arguments)
- Allows passing multiple keyword arguments as a dictionary.


        def info(**details):
             for key, value in details.items():
                 print(f"{key}: {value}")

        info(name="Dikshant", age=23, city="Pune")

        Output:

        name: Dikshant
        age: 23
        city: Pune

**4. What is the purpose of the `return` statement in a Python function?**
- The "return" statement in Python is used to send a value from a function back to the caller. It serves the following purposes:  

   - Returns a Value – It allows a function to output a result that can be used later.  
   - Ends Function Execution – Once "return" is executed, the function stops running.  
   - Enables Reusability – Returned values can be stored in variables and used in other operations.  
   - Can Return Multiple Values – Python allows returning multiple values as tuples.  
   - Returns "None" by Default – If no "return" is specified, the function returns "None".  

Example:

           def square(num):
               return num * num

           result = square(4)
          print(result)  # Output: 16
          # Without "return", the function would not provide an output that can be stored or used later.

**5. What are iterators in Python and how do they differ from iterables?**
   - An iterable is any object that can be looped over (e.g., using a `for` loop). It contains elements that can be accessed one by one.  
- Examples: Lists, tuples, dictionaries, sets, strings.  
- It must have the `__iter__()` method, which returns an iterator.  


- An iterator is an object that produces elements one at a time using the `__next__()` method. It keeps track of its current position and does not restart automatically.  
  - It must implement both `__iter__()` and `__next__()` methods.  
  - When `__next__()` reaches the end, it raises a `StopIteration` error.  


Example:


      numbers = [1, 2, 3]   # Iterable (List)
      num_iter = iter(numbers)  # Creating an Iterator from Iterable
      print(next(num_iter))  # Output: 1  # Accessing elements using next()
      print(next(num_iter))  # Output: 2
      print(next(num_iter))  # Output: 3  
      # next(num_iter)  
      # Raises StopIteration

 **6. Explain the concept of generators in Python and how they are defined?**
   - A generator is a special type of iterator that allows you to iterate over a sequence of values, but it does not store all values in memory at once. Instead, it generates values on-the-fly when needed, making it memory efficient, especially for large datasets.  

   - How Generators Work :
- Generators produce values one at a time and only when requested (lazy evaluation).  
- They use the "yield" keyword instead of "return" to produce a value. When "yield" is called, the function’s state is saved, and the next time the generator is called, it resumes from where it left off.

   - How to Define a Generator :  
     - Generators can be defined in two ways:  
       - Using a generator function(with "yield").
       - Using a generator expression(a concise form similar to list comprehensions).

   - Generator Function :
    - A function that uses "yield" to return values.  
- The function doesn't return all values at once. It "yields" one value and pauses until the next value is requested.  

 Example :

          def count_up_to(limit):
              count = 1
              while count <= limit:
                   yield count
                   count += 1

          gen = count_up_to(3)
          for num in gen:
               print(num)
         # Output: 1
                   2
                   3
         # Each time `yield` is called, the function pauses, and when the next value is needed, the function resumes from that point.

- Generator Expression :
    - A more compact version of a generator defined directly in an expression. It works like a list comprehension but returns an iterator instead of a list.

Example :

          gen = (x * x for x in range(1, 4))
          for num in gen:
                print(num)

          # Output: 1
                    4
                    9


- Benefits of Generators:
   - Memory Efficient: They don’t store all values in memory. This is particularly useful when working with large datasets.
   - Lazy Evaluation: Values are generated only when needed, making generators more efficient for iterating over large sequences.
   - Cleaner Code: They can simplify code for complex iteration logic without requiring explicit state management.

**7. What are the advantages of using generators over regular functions?**
   - Memory Efficiency
      - Generators produce values one at a time and do not store the entire sequence in memory, which makes them memory-efficient, especially for large datasets or infinite sequences.
      - Regular functions (that return a list, for example) store all values in memory, which can lead to higher memory usage, especially when dealing with large amounts of data.

- Lazy Evaluation  
   - Generators use lazy evaluation, meaning they only generate the next value when it’s requested. This can speed up the initial execution and reduce unnecessary computations.
   - Regular functions that return lists or other collections compute and store all values upfront, even if they are not all needed immediately.

- Cleaner Code and More Readable
   - Generators often lead to cleaner and more readable code for iterative processes. With "yield", you don’t need to manually manage iteration state (like maintaining a counter or manually managing an iterator).
   - Regular functions may require explicit handling of state, which can make code longer and harder to maintain.

- Improved Performance with Large Data
   - Generators can process large datasets efficiently because they don’t require the whole dataset to be loaded into memory at once. This is particularly useful when working with streams, file processing, or large database queries.
   - Regular functions that return complete collections might be slow and inefficient when dealing with large volumes of data due to the need to load all values into memory at once.

- Can Represent Infinite Sequences
   - Generators can easily represent infinite sequences or large ranges of data (such as counting numbers or reading large files line by line) without worrying about memory overflow, since values are generated on demand.
   - Regular functions would struggle to represent infinite or unbounded sequences without storing them entirely in memory.

- Better Handling of Infinite or Long Sequences
   - Generators allow you to iterate through sequences that would be impossible to fit in memory (e.g., an infinite sequence). They can stop at any point, yielding results on the fly without exhausting memory.
   - Regular functions would require storing the entire sequence in memory, which isn't feasible for infinite or extremely long sequences.

**8. What is a lambda function in Python and when is it typically used?**
   - A lambda function is a small, anonymous function defined with the lambda keyword, rather than the def keyword. It can have any number of input parameters but only a single expression. The result of that expression is implicitly returned.

            Syntax:
                lambda arguments: expression
                # arguments: The input parameters to the function.
                # expression: The single expression whose result is returned.

When to Use Lambda Functions:
- For short, throwaway functions: When you need a function temporarily and don’t want to write a full function definition.
- When you need a simple function as an argument: Often used with functions like map(), filter(), sorted(), or reduce() where a small function is needed for a short operation.
- For functional programming styles: In cases where you prefer to use higher-order functions and don’t want to define a full function with def.

Examples :

- Basic Example of a Lambda Function

       add = lambda x, y: x + y
       print(add(5, 3))  # Output: 8
       # Here, add is a lambda function that takes two arguments and returns their sum.

- Using Lambda with map()
  - map() applies a function to every item of an iterable.

            numbers = [1, 2, 3, 4]
            squared = list(map(lambda x: x**2, numbers))
            print(squared)  # Output: [1, 4, 9, 16]
            # In this case, the lambda function is used to square each element in the list.

- Using Lambda with filter()
  - filter() filters the elements based on a condition provided by the function.

           numbers = [1, 2, 3, 4, 5, 6]
           even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
           print(even_numbers)  # Output: [2, 4, 6]
           #Here, the lambda function filters out the even numbers from the list.

- Using Lambda with sorted()
   - We can use a lambda function to specify sorting logic.

           data = [(1, 'apple'), (3, 'banana'), (2, 'cherry')]
           sorted_data = sorted(data, key=lambda x: x[1])
           print(sorted_data)  # Output: [(1, 'apple'), (3, 'banana'), (2, 'cherry')]
           #In this example, the lambda function sorts the list of tuples based on the second element (the string).

**9. Explain the purpose and usage of the `map()` function in Python?**
   - The map() function in Python is used to apply a given function to all items in an iterable (such as a list, tuple, etc.) and return a new iterable (a map object) that contains the results.

      - Syntax:

       map(function, iterable)

       #function: The function that will be applied to each item in the iterable.
       #iterable: The iterable whose elements are to be processed by the function.

   - When to Use map():
- When you need to apply a function to every element in an iterable.
- When working with transformations on data, such as modifying or filtering items in a list or other iterable.
- When you prefer functional programming approaches (where operations are treated as functions applied to data).

Example : Applying a Function to Each Element in a List
                    
                    Suppose we want to square each number in a list.

                         numbers = [1, 2, 3, 4]
                         squared = list(map(lambda x: x ** 2, numbers))
                        print(squared)  # Output: [1, 4, 9, 16]
                    #In this example, the lambda function squares each element in the numbers list.

Example : Converting Strings to Uppercase

                    words = ['hello', 'world', 'python']
                    uppercased = list(map(str.upper, words))
                    print(uppercased)  # Output: ['HELLO', 'WORLD', 'PYTHON']
                    #Here, the map() function applies the str.upper method to each string in the words list to convert them to uppercase.

   Example : Adding Corresponding Elements of Two Lists
  - We can also use map() with multiple iterables to apply a function to corresponding elements from each iterable.

                      list1 = [1, 2, 3]
                      list2 = [4, 5, 6]
                      sum_lists = list(map(lambda x, y: x + y, list1, list2))
                      print(sum_lists)  # Output: [5, 7, 9]
                      #In this case, the lambda function adds the corresponding elements from list1 and list2.

  - Benefits of Using map():
- Concise and Clean Code: Instead of writing loops to apply a function to each item, map() provides a more compact way to do it.
- Better for Functional Programming: It aligns with a functional programming style, where you apply functions to data rather than using traditional for-loops.
- Performance: Since it returns an iterator, map() can be more efficient for large datasets, as it doesn’t generate the whole result at once.

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**
   

| Feature       | **`map()`** | **`filter()`** | **`reduce()`** |
|--------------|------------|---------------|---------------|
| **Purpose**  | Applies a given function to each element of an iterable | Filters elements from an iterable based on a condition | Reduces an iterable to a single cumulative value |
| **Input**    | A function and one or more iterables | A function that returns `True` or `False` and an iterable | A function that takes two arguments and an iterable |
| **Output**   | An iterator (map object) containing transformed values | An iterator (filter object) containing only elements that satisfy the condition | A single accumulated result |
| **Function Type** | Transformation | Filtering | Aggregation |
| **Return Type** | `map` object (needs conversion to list/tuple for viewing) | `filter` object (needs conversion to list/tuple for viewing) | A single value |
| **When to Use** | When you need to modify or transform each element in an iterable | When you need to keep only certain elements based on a condition | When you need to combine all elements into one result |
| **Common Use Cases** | Applying mathematical operations, converting strings, modifying data structures | Selecting even/odd numbers, filtering words based on length, removing None values | Summing a list, finding the maximum product, cumulative calculations |
| **Example** | `map(lambda x: x**2, [1,2,3,4]) → [1, 4, 9, 16]` | `filter(lambda x: x % 2 == 0, [1,2,3,4]) → [2, 4]` | `reduce(lambda x, y: x + y, [1,2,3,4]) → 10` |
| **Requires `functools` Module?** | No | No | Yes (`from functools import reduce`) |
| **Performance Consideration** | Efficient for element-wise transformations | More efficient than list comprehensions when filtering | Can be slow for large datasets as it processes sequentially |
| **Alternative Approach** | List comprehension: `[x**2 for x in numbers]` | List comprehension with condition: `[x for x in numbers if x % 2 == 0]` | Using built-in functions like `sum()` or `max()` |


- Use **`map()`** when you want to apply a function to **all** elements in an iterable.
- Use **`filter()`** when you want to **select** elements that meet a certain condition.
- Use **`reduce()`** when you need to **combine** all elements into a single result.

```python
# Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_evens(numbers):
    evens = []
    for num in numbers:
        if num % 2 == 0:
            evens.append(num)
    return sum(evens)

numbers = [1, 2, 4, 6, 9, 12]
print(sum_of_evens(numbers))

```

```python
# Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            self.count += 1
            self.a, self.b = self.b, self.a + self.b
            return self.a
        raise StopIteration

num_terms = int(input("Enter number of terms: "))
fibonacci = FibonacciIterator(num_terms)

for num in fibonacci:
    print(num, end=" ")


```

```python
# Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_2(exp):
    for i in range(exp + 1):
        yield 2 ** i

exp = int(input("Enter the exponent: "))
for power in powers_of_2(exp):
    print(power, end=" ")
```

```python
# Q7.  Implement a generator function that reads a file line by line and yields each line as a string.

import requests

def read_file_line_by_line(url):
    response = requests.get(url) # Send a GET request to the URL to fetch the content

    for line in response.text.splitlines(): # Split the text by lines and yield each line
        yield line

url = 'https://people.sc.fsu.edu/~jburkardt/data/csv/addresses.csv'
for line in read_file_line_by_line(url):
    print(line)
```

```python
# Q10.  Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(s):
    vowels = "aeiouAEIOU"

    filtered_chars = filter(lambda char: char not in vowels, s)

    result = ""
    for char in filtered_chars:
        result += char
    return result

input_string = "Dikshant Bhoyar"
output_string = remove_vowels(input_string)


print(output_string)
```
